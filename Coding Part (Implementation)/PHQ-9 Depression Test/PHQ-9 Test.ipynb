```
# import the library
from nltk.translate.bleu_score import sentence_bleu
from textblob import TextBlob
from nltk.translate.bleu_score import corpus_bleu
from nltk.translate.bleu_score import SmoothingFunction
cc = SmoothingFunction()
```

```
# Options
# 1. Not at all 
# 2. Several Days
# 3. More than half of the days
# 4. Nearly Every Day
```

```
reference_1 = [[["not", "at", "all"], ["no"], ["enough"], ["sufficient"], ["never"]]]
reference_2 = [[["several", "days"],["few", "days"], ["sometime"], ["rarely"], ["barely"]]]
reference_3 = [[["more", "than", "half", "of", "the", "day"], ["mostly"], ["many", "times"], ["very", "frequently"], ["very", "often"], ["many", "of", "the", "days"]]]
reference_4 = [[["everyday"], ["each", "day"], ["moderate"], ["yes"], ["daily"], ["everytime"], ["more", "than", "enough"],["alot"], ["all", "time"], ["always"], ["yeah"]]]
```

```
# create a function to pre-process the answer
def pre_process_answer(answer):

  # lower all the characters
  answer = answer.lower()

  # split the answer into words
  words = (answer.split())

  # print the word for checking purpose
  # test pass
  # print(words)

  if len(words)==1:
    words.insert(0,"a")

  return words
```

```
# create a function to calculate the bleu score
def bleu_score(candidate):

  # initialize score
  score = 0

  # calculate score for rating 1, 2, 3 and 4.
  score_1 = corpus_bleu(reference_1, candidate, smoothing_function=cc.method4)
  score_2 = corpus_bleu(reference_2, candidate, smoothing_function=cc.method4)
  score_3 = corpus_bleu(reference_3, candidate, smoothing_function=cc.method4)
  score_4 = corpus_bleu(reference_4, candidate, smoothing_function=cc.method4)

  # add respective rating to score
  if score_1>score_2 and score_1>score_3 and score_1>score_4:
    score = 0
  elif score_2>score_1 and score_2>score_3 and score_2>score_4:
    score = 1
  elif score_3>score_1 and score_3>score_2 and score_3>score_4:
    score = 2
  elif score_4>score_1 and score_4>score_2 and score_4>score_3:
    score = 3 

  return score
```

```
# create a function to calculate the bleu score
def calculate_bleu_score(words):

  # store the words into temporary variable
  candidate = [words]

  # print all the words for checking purpose
  # test pass
  # print(candidate)

  #if len(words)==1:
  #  score = sentiment_analysis(words)

  #else:
  score = bleu_score(candidate)

  # return score
  return score  
```

```
# create a function to ask the questionnarie
def questionnaire_score(answer):

  # create a list of words from answers
  words = pre_process_answer(answer)

  # calculate the final score
  score = calculate_bleu_score(words)

  return score
```

```
# All questions
Q_1 = "How often have you don't found interest or pleasure in doing things?\n"
Q_2 = "\nAre you feeling down or depressed or hopeless?\n"
Q_3 = "\nHow often do you feel sleepy?\n"
Q_4 = "\nHow often do you feel tired? Do you feel lethargic after eating?\n"
Q_5 = "\nHow much is your food intake? How much do you eat everyday?\n"
Q_6 = "\nHow often do you feel yourself being responsible for everything happening around you?\n"
Q_7 = "\nDo you find it difficult to concentrate on your studies?\n"
Q_8 = "\nDo you feel anxious while talking to people?\n"
Q_9 = "\nHow many times have you thought of taking your own life or ending yourself?\n"

# create a list of questions
list_questions = [Q_1, Q_2, Q_3, Q_4, Q_5, Q_6, Q_7, Q_8, Q_9]
```

```
# final score
score = 0

for question in list_questions:
  # take the answer from user
  answer = input(question)

  # calculate the score
  print(questionnaire_score(answer))
  score += questionnaire_score(answer)

print("Depression Score :- ",score)
print("Depression Level :- ",end="")

# get the stage of the depression
if score>=1 and score<=4:
  print("Minimal Depression")
elif score>=5 and score<=9:
  print("Mild depression")
elif score>=10 and score<=14:
  print("Moderate Depression")
elif score>=15 and score<=19:
  print("Moderate severe Depression")
elif score>=20 and score<=27:
  print("Severe Depression")
```
