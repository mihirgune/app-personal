```
# import json and load the file in json format
import json
path = "/content/drive/My Drive/2020_Intern_03_VIIT_03_Chatbot/Coding Part (Implementation)/Database and JSON/All Data/Main Data/1. COVID-QA.json"
file = open(path)
```

```
# load the data in json format
data = json.load(file)
```

```
# test the data is load or not
print(len(data['data']))
```

```
# create a list of questioN and answer
question_list = []
answer_list = []

# loop over the data
for article_id in (range(len(data['data']))):
    list_paragraphs = data['data'][article_id]['paragraphs']
                
    # loop over the paragraphs
    for paragraph in list_paragraphs:
        context = paragraph['context']
    
        # clean the text
        qas = paragraph['qas']
                    
        # loop over Q/A
        for qa in qas:
            question = qa['question']
            question_list.append(question)
            
            answer_ids = 1 if qa['answers'] else 0

            for answer_id in range(answer_ids):
                answer = qa['answers'][answer_id]['text']
                answer_list.append(answer)
```

```
# check for the lenght of question list
print(len(question_list))
```

```
# check for the length of answer list
print(len(answer_list))
```

```
# test whether the questions are in the list or not
print(question_list) 
```

```
# test whether the answers are in the list or not
print(answer_list)
```

```
# use pandas to create dataframe
import pandas as pd
```

```
# get the list of tuples from two lists and merge them by using zip().  
list_of_tuples = list(zip(question_list, answer_list))
```

```
# Converting lists of tuples into pandas Dataframe.  
df = pd.DataFrame(list_of_tuples, columns = ['Questions', 'Answers'])
```

```
# check the questions and answers
df.head()
```

```
# convert this file to csv file
df.to_csv('/content/drive/My Drive/2020_Intern_03_VIIT_03_Chatbot/Coding Part (Implementation)/Database and JSON/All Data/QA List Data/COVID_QA.csv')
```
