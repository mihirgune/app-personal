```
# import json and load the file in json format
import json
path = "/content/drive/My Drive/2020_Intern_03_VIIT_03_Chatbot/Coding Part (Implementation)/Database and JSON/All Data/Main Data/2. Basic Conversation.json"
file = open(path)
```

```
# load the data in json format
data = json.load(file)
```

```
# check the data is loaded or not
print(data)
```

```
# check for the lenght of data
print(len(data))
```

```
# APPROACH ONE
# THIS APPROACH IS FOR GETTING BETTER KNOWLEDGE OF DATASET

# add questions to question list
question_list = []

# store the values of questions
questions = data.values()

# iterate over each value in answer
for value in questions:
    context = value['in_response_to']
    
    # store the question which having text as a key
    for qa in context:
        question = qa['text']

        if questions:
            question_list.append(question)

# check the list
print(question_list)
```

```
# THIS APPROACH IS FOR GETTING BETTER KNOWLEDGE OF DATASET

# add answers to question list
answer_list = []

answers = data.keys()

for answer in answers:
    answer_list.append(answer)

print(answer_list)
```

```
# check the length of the answer list processed above
print(len(answer_list))
```

```
# check the length of the question list processed above
print(len(question_list))
```

```
# As we can see that, length of answers list and questions list is not matching. 
# Hence we need to pre-processed the data once again and delete those answers which are ask for empty list
# Example I found is mentioned below :- 
# {'Can I ask you a question?': {'in_response_to': []}}
```

```
# APPROACH TWO :- USE THE DICTIONARY KEY, VALUE PAIR

# data preprocessing
questions_list = []
answers_list = []

# iterate over each key and values pair
for key,value in data.items():

    # question is present
    if value['in_response_to']:

        # save the question dictionary to context
        context = value['in_response_to']

        # append the question and answer to question_list and answer_list
        for temp in context:
            questions_list.append(temp['text'])
            answers_list.append(key)
```

```
# check the length of the answer list processed above
print(len(answers_list))
```

```
# check the length of the question list processed above
print(len(questions_list))
```

```
# Now we got same lengths i.e., we have same number of questions and for each question there is answer present.
```

```
# check the questions list
print(questions_list)
```

```
# check the answers list
print(answers_list)
```

```
# we can see clearly see that all the data preprocessed and we get appropriate answer for each question.
# Pre-processing for 2. Normal Conversation is DONE!!!
```

```
# use pandas to create dataframe
import pandas as pd
```

```
# get the list of tuples from two lists and merge them by using zip().  
list_of_tuples = list(zip(questions_list, answers_list))
```

```
# Converting lists of tuples into pandas Dataframe.  
df = pd.DataFrame(list_of_tuples, columns = ['Questions', 'Answers'])
```
