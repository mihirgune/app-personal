```
# importing pandas to load and do operations on our data
import pandas as pd
```

```
# load the dataset
path = "/content/drive/My Drive/2020_Intern_03_VIIT_03_Chatbot/Coding Part (Implementation)/Database and JSON/All Data/Main Data/3. Mental Health Queries.xlsx"
df = pd.read_excel(path)
```

```
# drop the unwanted columns
# cols :- Unnamed: 0, questionText, questionLink,	therapistInfo, therapistURL, split
df = df.drop(['Unnamed: 0', 'questionText', 'questionLink', 'therapistInfo', 'therapistURL', 'split'], axis = 1)
```

```
# As per the obervations
# Many of the questions are repeated and scattered appart.
# We cannot assign question ID's to such questions
# Sort the dataframe
df = df.sort_values(by=['questionTitle'])
```

```
# check for sorting is done properly or not.
df.head()
```

```
# we can see above that index of dataframe is not sorted hence we have to reset all the indices.
df = df.reset_index(drop=True)
df.head()
```

```
# check of the base output
question = df['questionTitle'][0]
print(question)
```

```
# processed the question-ID data
row = 0
rows = df.shape[0]
ID = 0
question = df['questionTitle'][0]

while row<rows:
    
    if df.iloc[row]['questionTitle'] != question:
        ID += 1
        question = df['questionTitle'][row]
        row -= 1
    else:
        df['questionID'][row] = ID

    row += 1
```

```
# check for the process we have done earlier
df.head()
# Getting correct OUTPUT as per expectation
```

```
# Find the max upvotes value for each answer
ID = 0
rows = df.shape[0]
row = 0
questions_list = []
answers_list = []
topics_list = []
count = 0
start_index = 0
end_index = 0

while row<rows:
    
    if df.iloc[row]['questionID'] != ID:
        end_index += count
        main_index = 0

        # find the max upvote value from start index to end index
        max = -1
        for i in range(start_index,end_index):
            value = df.iloc[i]['upvotes']
            if value>max:
                max = value
                main_index = i

        # lets check that we get zero as a max values for upvotes
        if max == 0:

            # check for maximum views
            views_max = -1
            for i in range(start_index,end_index):
                value = df.iloc[i]['views']
                if value>views_max:
                    views_max = value
                    main_index = i

        # append the question and its best answer
        questions_list.append(df.iloc[main_index]['questionTitle'])
        answers_list.append(df.iloc[main_index]['answerText'])

        # append only unique topic
        topic = df['topic'][main_index]
        if topic not in topics_list:
            topics_list.append(topic)

        # reset all the numbers
        ID += 1
        row -= 1
        start_index += count
        count = 0

    else:
        # count the number of common questions
        count += 1

    # increament row to terminate the while loop 
    row += 1
```

```
# get the list of tuples from two lists and merge them by using zip().  
list_of_tuples = list(zip(questions_list, answers_list))
```

```
# Converting lists of tuples into pandas Dataframe.  
df = pd.DataFrame(list_of_tuples, columns = ['Questions', 'Answers'])
```
